# Despliegue seguro en Linux.

Este cuaderno lleva seguridad y entornos al terreno operativo de servidores Linux: acceso remoto, endurecimiento, promocion por ambiente y control de cambios.

## Objetivos.

- Entender por que SSH y secretos requieren tratamiento estricto.
- Preparar despliegues por ambiente.
- Conectar pipeline y operacion segura.

## Controles minimos.

Para un despliegue serio en Linux se requiere al menos:

- autenticacion remota acotada y rotacion de credenciales;
- separacion por ambientes;
- secretos fuera del repositorio;
- pasos de despliegue idempotentes;
- evidencia posterior al cambio.

## Ejemplo conceptual: despliegue remoto.

```yaml
- name: Configure SSH
  run: |
    install -m 700 -d ~/.ssh
    echo "$SSH_PRIVATE_KEY" > ~/.ssh/id_ed25519
    chmod 600 ~/.ssh/id_ed25519
    ssh-keyscan -H "$DEPLOY_HOST" >> ~/.ssh/known_hosts

- name: Deploy to Linux server
  run: |
    scp docker-compose.yml "$DEPLOY_USER@$DEPLOY_HOST:/srv/app/"
    ssh "$DEPLOY_USER@$DEPLOY_HOST" "cd /srv/app && docker compose pull && docker compose up -d"
```

El punto didactico es que SSH no solo transporta comandos: formaliza un cambio operativo sobre un host real y por eso debe estar sujeto a los mismos controles que el resto del pipeline.

## Flujo de despliegue por SSH.

Cuando el destino es un servidor Linux tradicional, el pipeline suele necesitar una ruta de despliegue remoto por SSH. Esa ruta no debe tratarse como un simple `ssh user@host comando`, sino como una operacion controlada con identidad, host verification, secretos acotados y pasos idempotentes.

Un flujo sano suele incluir:

- preparacion del host con usuarios y permisos minimos;
- distribucion segura de la clave o uso de certificados temporales;
- verificacion del `known_hosts`;
- copia o actualizacion del artefacto;
- reinicio controlado del servicio;
- validacion posterior al despliegue.
