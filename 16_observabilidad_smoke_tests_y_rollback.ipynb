# Observabilidad, smoke tests y rollback.

Py271 trata la observabilidad como parte del sistema de entrega y no como una tarea posterior del equipo de operaciones. Rollback se estudia como respuesta guiada por señales, no como improvisacion.

## Objetivos.

- Diseñar verificaciones posteriores al deploy.
- Distinguir logs, metricas, trazas y alertas como señales complementarias.
- Relacionar smoke tests con liberacion segura.
- Entender rollback como capacidad y no como improvisacion.

## Verificar antes de dar por bueno el despliegue.

Un smoke test comprueba que el servicio recien desplegado responde en su ruta minima esperada. Esa verificacion debe ir acompañada de observabilidad basica: logs, metricas, estado del proceso, latencia, tasa de errores y señales de saturacion.

La idea central es simple: si el pipeline despliega pero nadie puede observar el estado real del servicio, todavia no existe entrega defendible. Rollback no es solo volver a una version anterior; es conservar trazabilidad suficiente para saber a que version volver, bajo que criterio hacerlo y como validar que la recuperacion funciono.

## Pila minima de observabilidad.

En un curso de CI/CD no hace falta cubrir una plataforma completa de observabilidad, pero si es necesario fijar el modelo operativo minimo:

- logs estructurados para reconstruir eventos y errores;
- metricas de salud como latencia, throughput y error rate;
- trazas cuando un request atraviesa varios componentes;
- alertas que traduzcan desviaciones tecnicas en accion operativa.

Una entrega continua madura define desde antes del despliegue que señales se observaran, con que umbrales y quien responde cuando algo sale del rango esperado.

## SLI, SLO y alertas.

Los smoke tests sirven para responder 'arranco o no arranco', pero la observabilidad responde 'esta funcionando dentro de parametros aceptables'. Por eso conviene introducir al menos estas ideas:

- SLI: una medicion observable, por ejemplo tasa de error o latencia p95;
- SLO: el objetivo esperado para ese indicador;
- alerta: la regla que dispara investigacion o rollback cuando el servicio se aparta del objetivo.

## Aplicacion a las rutas del curso.

La observabilidad no se revisa igual en todos los destinos de despliegue. En Py271 conviene contrastar las dos rutas principales:

- Google Cloud Functions: revisar invocaciones, errores, duracion, logs estructurados y comportamiento por version desplegada;
- Linux por SSH: revisar logs de la aplicacion, proxy inverso, procesos o contenedores, consumo de CPU y memoria, puertos expuestos y estado del servicio.

Eso permite mostrar que el mismo pipeline termina en entornos distintos, pero ambos necesitan evidencia operativa despues del deploy.

## Runbook minimo posterior al despliegue.

Un runbook breve pero util deberia responder al menos estas preguntas:

- que panel, comando o fuente de logs se consulta primero;
- que umbral obliga a investigar y cual obliga a revertir;
- quien aprueba el rollback en `prod`;
- como se valida que la version recuperada quedo sana.

Sin ese runbook, la observabilidad queda como telemetria pasiva y no como capacidad real de operacion.

```python
indicadores = [
    'healthcheck',
    'smoke tests',
    'logs estructurados',
    'latencia p95',
    'error rate',
    'trazas',
    'alertas',
    'criterio de rollback',
]

for indicador in indicadores:
    print('control:', indicador)
```
