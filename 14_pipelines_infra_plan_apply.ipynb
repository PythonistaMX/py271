# Pipelines de infraestructura: plan y apply.

Aqui convergen dos lineas del curso: la estructura de workflows y la disciplina de Terraform introducida en Py271.

## Objetivos.

- Integrar fmt, validate, plan y apply en un workflow.
- Entender aprobaciones por entorno.
- Separar validacion automatica de cambios destructivos.

## Dos fases distintas.

Una practica sana separa `plan` y `apply`. La primera es automatizable casi siempre; la segunda requiere control de cambios, aprobaciones o restriccion por rama y entorno.

Las reglas de proteccion de entornos son especialmente utiles aqui porque ayudan a decidir quien puede promover un cambio de infraestructura.

## Ejemplo concreto: plan en pull request, apply en main.

Una politica tipica de infraestructura es esta:

- en `pull_request`: ejecutar `fmt`, `validate` y `plan`;
- en `push` a `main`: permitir `apply`;
- en `prod`: exigir aprobacion manual via entorno protegido.

```yaml
on:
  pull_request:
  push:
    branches: ["main"]
```

Ese diseno une workflow de repositorio, entorno y politica de cambio de infraestructura en una sola ruta operativa.

````python
etapas = ['fmt', 'validate', 'plan', 'apply']

for etapa in etapas:
    print('pipeline IaC:', etapa)
````
