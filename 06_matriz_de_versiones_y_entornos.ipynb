# Matriz de versiones y entornos.

Esta sesion mezcla runners, artefactos y entornos. En Py271 se estudian juntos porque un pipeline real debe decidir donde corre, que conserva y bajo que reglas promueve cambios.

## Objetivos.

- Relacionar runners efimeros con artefactos persistentes.

## Runners, artefactos y entornos.

Los runners hospedados por GitHub son entornos efimeros. Esa caracteristica vuelve util la `strategy.matrix`: permite validar el mismo cambio en varias versiones sin duplicar jobs completos.

Los artefactos existen porque el runner desaparece al terminar el job. Tambien trabajamos la nocion de entornos protegidos: desarrollo, pruebas, staging y produccion pueden tener variables, secretos, aprobaciones y ventanas de despliegue distintas.

![image.png](img/12/12-01.png)

## Que es un entorno en GitHub Actions.

Un entorno representa un destino o fase del ciclo de entrega, por ejemplo `dev`, `qa`, `staging` o `prod`. No es solo una etiqueta: es un contenedor de reglas operativas que GitHub aplica antes de permitir que un job sensible continúe.

Un entorno puede concentrar:

- variables y secretos especificos;
- aprobaciones manuales por personas autorizadas;
- restricciones por rama;
- ventanas de espera antes del despliegue.

Esto es util porque separa el pipeline tecnico de las politicas de promocion. El workflow define como desplegar; el entorno define bajo que controles puede hacerlo.

## Relacion entre job y entorno.

Un job se asocia a un entorno mediante la clave `environment`. Desde ese momento, GitHub puede exigir aprobaciones, exponer `vars` y `secrets` del entorno correcto y registrar mejor la trazabilidad del despliegue.

```yaml
jobs:
  deploy:
    runs-on: ubuntu-latest
    environment: prod
    steps:
      - uses: actions/checkout@v4
      - run: echo "Desplegando a produccion"
```

Este modelo evita mezclar todos los secretos del repositorio en un solo contexto y ayuda a que `apply` o `deploy` no se ejecuten con la misma libertad que un job de lint o pruebas.

## Ejemplo concreto: dev, test y prod.

Un mismo artefacto puede recorrer tres entornos con controles distintos. Por ejemplo:

- `dev`: despliegue automatico al hacer merge a `main`;
- `test`: despliegue posterior a smoke tests y validacion funcional;
- `prod`: despliegue sujeto a aprobacion manual y observabilidad reforzada.

```yaml
jobs:
  deploy_dev:
    environment: dev
  deploy_test:
    environment: test
  deploy_prod:
    environment: prod
```

La clave no es cambiar el artefacto, sino cambiar el nivel de control operativo de cada entorno.

````python
python_versions = ['3.10', '3.11', '3.12']
entornos = ['dev', 'qa', 'prod']

for version in python_versions:
    print('validar en', version)
print('promocion posible hacia:', entornos)
````
