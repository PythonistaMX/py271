# Introduccion a CI/CD y GitHub Actions.

Este cuaderno presenta la base conceptual de GitHub Actions y la ubica en Py271 dentro de la serie py2xx.

## Objetivos.

- Entender que problema resuelve CI/CD.
- Identificar la anatomia basica de GitHub Actions.
- Preparar la base para quality gates, IaC y despliegue seguro.

## Conceptos base.

Durante el ciclo de vida del software aparecen tareas repetitivas, lentas y susceptibles de error humano. En Py271 esto se traduce en una regla operativa: todo paso manual que agregue riesgo pero no valor debe ser candidato a automatizacion.

GitHub Actions permite reaccionar ante eventos del repositorio y ejecutar workflows sobre runners efimeros. Cada workflow contiene jobs y cada job contiene steps. Esa anatomia sera la base para incorporar seguridad, contenedores, Terraform y despliegue seguro.
