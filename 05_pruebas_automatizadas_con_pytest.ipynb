# Pruebas automatizadas con pytest.

En Py271 las pruebas automatizadas dejan de ser un ejemplo accesorio y se convierten en una condicion minima para empaquetar y desplegar.

## Objetivos.

- Integrar pytest en un job de CI.
- Relacionar pruebas, checks y reglas de merge.
- Entender el valor de un entorno efimero para validacion.

## El valor de probar en runners efimeros.

Cada ejecucion del pipeline parte de un entorno limpio. Eso obliga a declarar dependencias, datos y pasos de preparacion de forma explicita. La ventaja es que reduce diferencias entre lo que funciona localmente y lo que realmente es desplegable.

Un job de pruebas bien construido no solo ejecuta `pytest`; tambien deja evidencia clara en los checks del pull request y ayuda a decidir si un merge es aceptable.

```python
resultado = {'tests': 12, 'passed': 12, 'failed': 0}
estado = 'verde' if resultado['failed'] == 0 else 'rojo'

print('Resumen:', resultado)
print('Pipeline:', estado)
```
