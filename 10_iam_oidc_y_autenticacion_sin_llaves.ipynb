# IAM, OIDC y autenticacion sin llaves.

Este cuaderno introduce OpenID Connect como alternativa a secretos de larga duracion. Py271 lo convierte en pieza central para autenticar pipelines contra GCP.

## Objetivos.

- Entender OIDC en GitHub Actions.
- Relacionar federacion con minimo privilegio.
- Evitar credenciales estaticas en el pipeline.

## OIDC como control de seguridad.

Si un workflow necesita acceder a recursos de nube compatibles con OpenID Connect, es preferible autenticarlo mediante federacion en lugar de guardar llaves estaticas como secretos.

En Py271 eso significa intercambiar identidad efimera por permisos de corta duracion y acotados a un service account con privilegios minimos.

## Ejemplo concreto: identidad por repositorio y entorno.

Un diseño sano evita una sola credencial global para todo. En su lugar, el repositorio y el entorno ayudan a decidir que identidad puede solicitar el pipeline.

Por ejemplo, un job de `deploy_dev` puede autenticar contra un service account con permisos limitados, mientras `deploy_prod` usa otro con alcance menor y sujeto a aprobacion previa.

Eso alinea tres capas a la vez: repositorio protegido, entorno correcto e identidad efimera via OIDC.

```python
roles = [
    'artifactregistry.writer',
    'run.admin',
    'iam.serviceAccountUser',
]

for rol in roles:
    print('rol requerido:', rol)
```
