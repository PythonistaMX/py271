# YAML y sintaxis de workflows.

Este cuaderno desarrolla YAML, triggers, jobs, steps, expresiones y variables con el formato editorial de Pythonista.

## Objetivos.

- Entender la estructura minima de un workflow.
- Identificar los triggers principales.
- Reconocer el papel de expresiones y contextos.

## Estructura minima de un workflow.

Un workflow se define dentro de `.github/workflows/` y normalmente contiene bloques como `name`, `on`, `jobs` y `steps`. Tambien usaremos expresiones `${{ ... }}` y contextos como `github`, `env`, `vars` y `secrets`.

```yaml
name: ci
on:
  push:
    branches: ["main"]
  pull_request:
jobs:
  checks:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - run: echo ${{ github.ref }}
```

## Que es un workflow y como se lee.

Un workflow es la unidad de automatizacion de GitHub Actions. Se dispara cuando ocurre uno o varios eventos definidos en `on` y organiza el trabajo que debe ejecutarse en respuesta a esos eventos.

Conviene leer un workflow de arriba hacia abajo en este orden:

- `name`: etiqueta legible de la automatizacion.
- `on`: condiciones que la activan.
- `jobs`: trabajos independientes o encadenados.
- `runs-on`: tipo de runner requerido por cada job.
- `steps`: secuencia de acciones y comandos dentro del job.

La idea importante es que un workflow no es un script lineal grande. Es una declaracion de condiciones, dependencias y pasos de ejecucion. Eso permite reusar acciones, paralelizar trabajos y aplicar reglas distintas segun rama, evento o entorno.

## Triggers, contextos y expresiones.

Los triggers responden a eventos del repositorio, por ejemplo `push`, `pull_request`, `workflow_dispatch` o `schedule`. A partir de ese evento, GitHub expone datos en distintos contextos como `github`, `runner`, `env`, `vars`, `secrets` y `matrix`.

Las expresiones `${{ ... }}` permiten usar esos datos para tomar decisiones dentro del workflow. Por ejemplo, habilitar un despliegue solo si la rama es `main`, cambiar variables por entorno o construir una matriz de versiones.

Esa combinacion de evento, contexto y expresion es lo que vuelve a GitHub Actions un DSL operativo para CI/CD y no solo un archivo YAML descriptivo.

## Ejemplo concreto: workflow para pull request y rama principal.

En un repositorio protegido, un flujo comun es este: una rama corta dispara checks con `pull_request`; despues del merge a `main`, otro evento `push` habilita pasos de promocion o despliegue controlado.

```yaml
on:
  pull_request:
    branches: ["main"]
  push:
    branches: ["main"]
```

Este patron conecta directamente el workflow tecnico con la politica del repositorio: el pull request valida; la rama principal promueve.

````python
workflow = {
    'name': 'ci',
    'on': ['push', 'pull_request', 'workflow_dispatch'],
    'jobs': {'checks': {'runs-on': 'ubuntu-latest'}}
}

for clave, valor in workflow.items():
    print(clave, '=>', valor)
````
