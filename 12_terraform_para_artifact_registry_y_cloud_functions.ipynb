# Terraform para Artifact Registry y Google Cloud Functions.

Py271 formaliza el aprovisionamiento con Terraform para que el pipeline no dependa de infraestructura creada manualmente al desplegar hacia GCP.

## Objetivos.

- Modelar recursos base para una ruta de despliegue en GCP.
- Relacionar artefactos versionados con el entorno de ejecucion.
- Entender el papel de Google Artifact Registry y Google Cloud Functions.

## Recursos tipicos.

Una ruta tipica modela al menos:

- un repositorio de artefactos e imagenes en Google Artifact Registry;
- una cuenta de servicio con permisos acotados;
- una funcion en Google Cloud Functions con configuracion por entorno;
- variables y secretos separados para dev, test y prod.

El objetivo es que la entrega continua tenga un destino reproducible y no dependiente de configuracion manual.

## Relacion entre contenedores, funciones y registry.

Aunque Google Cloud Functions abstrae parte de la operacion, el pipeline sigue necesitando empaquetado reproducible, versionado y control de artefactos. Artifact Registry permite centralizar dependencias construidas en GCP y mantener trazabilidad entre commit, build y despliegue.
